### Mean wirelength
x: rent, y: size, colour: mean wirelength

```python
def get_wirelength_estimations(p, size):
    k = np.array(range(1,int(size**0.5)*2))
    y = 3-2*p
    g = 100
    f_k = g/((k)**y)
    mean_wirelength = (f_k*k).sum()/f_k.sum()
    total_wirelength = (f_k*k).sum()
    return mean_wirelength, total_wirelength
```

```python
np.zeros((2,5))
```

```python
sizes = np.logspace(2,5, 30, base = 10)
rent_exponents = np.arange(0.5,1,0.02)
mean_wirelenghts = np.zeros((len(sizes), len(rent_exponents)))
total_wirelengths = np.zeros((len(sizes), len(rent_exponents)))
for i, size in enumerate(sizes):
    for j, p in enumerate(rent_exponents):
        mean_wirelength, total_wirelength = get_wirelength_estimations(p, size)
        [i, j] = mean_wirelength
        total_wirelengths[i, j] = total_wirelength
```

```python
cm = plt.cm.get_cmap('gist_rainbow')
xy = range(20)
z = xy
sc = plt.scatter(rent_exponents_plot, sizes_plot, c=mean_wirelenghts_plot, norm=colors.LogNorm(), s=35, cmap=cm)
plt.colorbar(sc)
plt.yscale('log')
plt.title('Mean wirelength estimation')
plt.xlabel('rent exponent p')
plt.ylabel('size K')
plt.show()
```
