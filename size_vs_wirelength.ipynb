### Mean wirelength
x: rent, y: size, colour: mean wirelength

```python
def get_wirelength_estimations(p, size):
    k = np.array(range(1,int(size**0.5)*2))
    y = 3-2*p
    g = 100
    f_k = g/((k)**y)
    mean_wirelength = (f_k*k).sum()/f_k.sum()
    total_wirelength = (f_k*k).sum()
    return mean_wirelength, total_wirelength
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
blocks = 850000
clusters = 1
for i in range(10):
    blocks_per_cluster= blocks/clusters
    clusters*=4
    print(blocks_per_cluster)
    
```

```python
def get_clusters(n_blocks):
    ks = np.arange(math.ceil(math.log(n_blocks, 4))+1)
    blocks_per_clusters = np.array([n_blocks]* len(ks) ) / 4**ks
    clusters = []
    for k, blocks_per_cluster in zip(ks, blocks_per_clusters):
        n_clusters = 4**k
        mean = blocks_per_cluster
        a = math.floor(blocks_per_cluster)
        b = math.ceil(blocks_per_cluster)
        if a == b:
            clusters.append(
                {
                    a:n_clusters
                }
            )        
        else:
            x = (mean - b)/(a-b)
            clusters.append(
                {
                    a:x * n_clusters,
                    b:(1-x) * n_clusters
                }
            )
#     if 1 in clusters[-2]:
#         clusters[-1][1] = clusters[-1][1] - clusters[-2][1]
    return clusters
```

```python
def get_clusters_n(n_blocks):
    ks = np.arange(math.ceil(math.log(n_blocks, 4))+1)
    blocks_per_clusters = np.array([n_blocks]* len(ks) ) / 4**ks
    clusters = []
    for k, blocks_per_cluster in zip(ks, blocks_per_clusters):
        n_clusters = 4**k

        clusters.append(
            {
                blocks_per_cluster:n_clusters
            }
        )
    return clusters
```

```python
r=0.55
def get_wirelength(clusters, r):
    n_blocks = list(clusters[0])[0]
    wirelength_k0 = math.sqrt(n_blocks)
    total_wirelength = 0
    n_prev_wires=0
    for k, cluster_k in enumerate(clusters):
        k = k-1
        if k >=0:
            wirelength_k = wirelength_k0 / 2**k
        else:
            wirelength_k = 0
        n_all_wires = 0
        for n_blocks, n_clusters in cluster_k.items():
            n_pins = n_blocks**r
            n_all_wires += n_pins/2*n_clusters
        n_wires = n_all_wires - n_prev_wires
        n_prev_wires = n_all_wires
        total_wirelength += n_wires * wirelength_k
#         print(n_wires * wirelength_k)
    return total_wirelength
```

```python
sizes = np.logspace(math.log(10000, 10),math.log(1000000, 10), 1000, base = 10)
rent_exponents = np.arange(0,1,0.005)

#sizes = np.logspace(math.log(50000, 10),math.log(200000, 10), 1000, base = 10)
#rent_exponents = np.arange(0.4,0.6,0.0005)

mean_wirelenghts = np.zeros((len(sizes), len(rent_exponents)))
total_wirelengths = np.zeros((len(sizes), len(rent_exponents)))
for i, size in enumerate(sizes):
    for j, p in enumerate(rent_exponents):
        clusters = get_clusters(size)
        total_wirelength = get_wirelength(clusters, p)
        total_wirelengths[i, j] = total_wirelength
```

```python
sizes_plot = np.repeat(sizes, len(rent_exponents))
rent_exponents_plot = list(rent_exponents) * len(sizes)
total_wirelenghts_plot = total_wirelengths.flatten()
```

```python
iso_wirelength_lines = []
min_wirelength = total_wirelengths.min()
max_wirelength = total_wirelengths.max()
a, b = int(math.log(min_wirelength, 2)), int(math.log(max_wirelength, 2))
wirelength_line_values = np.logspace(a ,b , (b-a)*2 + 1, base = 2)
wirelength = wirelength_line_values[5]
for wirelength in wirelength_line_values:
    iso_wirelength_line = []
    for i, p in enumerate(rent_exponents):
        size = sizes[np.argmin([abs(x-wirelength) for x in total_wirelengths[:, i]])]
        iso_wirelength_line.append([p, size])
    iso_wirelength_lines.append(np.array(iso_wirelength_line))
```

```python
max_rent = lambda b: (math.log(b,3)-1)/math.log(b,3)
max_rents = [max_rent(size) for size in sizes]
```

```python
def trend_line(data):
    x = data[:,0]
    y = data[:,1]
    x_mean = x.mean()
    y_mean = y.mean()
    x_err = x-x_mean
    y_err = y-y_mean
    a = (x_err*y_err).sum()/ ((x_err)**2).sum()
    b = y_mean - x_mean*a
    first_point = [x[0], x[0]*a + b]
    last_point = [x[-1], x[-1]*a + b]
    
    #Error
    error = sum((y - (x*a + b))**2)/len(data)   
    
    return np.array([first_point, last_point]), a, b, error
```

```python
sizes = np.logspace(math.log(10000, 10),math.log(1000000, 10), 1000, base = 10)
rent_exponents = np.arange(0.0,1,0.005)

scaling_factors = np.zeros((len(sizes), len(rent_exponents)))
for i, size in enumerate(sizes):
    for j, p in enumerate(rent_exponents):
        scaling_factor = (4-4**p)*(size**p)
        scaling_factors[i, j] = scaling_factor
```

```python
sizes_plot = np.repeat(sizes, len(rent_exponents))
rent_exponents_plot = list(rent_exponents) * len(sizes)
scaling_factors_plot = scaling_factors.flatten()
```

```python
cm = plt.cm.get_cmap('Pastel1')
xy = range(20)
z = xy
plt.figure(figsize=(15,10))

sc = plt.scatter(rent_exponents_plot, sizes_plot, c=scaling_factors_plot, s=35, cmap=cm, norm=colors.LogNorm())
plt.colorbar(sc)
plt.yscale('log')

plt.title('Scaling factors')
plt.xlabel('rent exponent p')
plt.ylabel('size K')
plt.plot(max_rents, sizes, color='red')
plt.show()
```
